```python
# LONG RESULT
import json
import os

response_data = []

# change to your path of long_mis
json_directory = "/Users/haley3/Desktop/cs152/cs152bots/automated_model/model_outputs/gemini_outputs/long_mis"

# Loop over all files in the directory
for filename in os.listdir(json_directory):
    if filename.endswith(".json"): 
        file_path = os.path.join(json_directory, filename)
        with open(file_path, 'r') as file:
            data = json.load(file)
            response_json = json.loads(data['response'])
            
            response_data.append(response_json)

df1 = pd.DataFrame(response_data)

df1
```

```python
# SHORT NONMIS
import json
import os

response_data = []

json_directory = "/Users/haley3/Desktop/cs152/cs152bots/automated_model/model_outputs/gemini_outputs/short_nonmis"

# Loop over all files in the directory
for filename in os.listdir(json_directory):
    if filename.endswith(".json"): 
        file_path = os.path.join(json_directory, filename)
        with open(file_path, 'r') as file:
            data = json.load(file)
            response_json = json.loads(data['response'])
            
            response_data.append(response_json)

df4 = pd.DataFrame(response_data)

df4
```

```python
df1["true_label"] = "MISOGYNISTIC"
df2["true_label"] = "NON-MISOGYNISTIC"


import pandas as pd
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming df_misogynistic and df_non_misogynistic are your two DataFrames

# Combine both DataFrames
df_combined = pd.concat([df1[['true_label', 'label']], df2[['true_label', 'label']]])

# Create confusion matrix
cm = confusion_matrix(df_combined['true_label'], df_combined['label'], labels=['MISOGYNISTIC', 'NON-MISOGYNISTIC'])

# Plot the confusion matrix
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['MISOGYNISTIC', 'NON-MISOGYNISTIC'], yticklabels=['MISOGYNISTIC', 'NON-MISOGYNISTIC'])
plt.title('Gemini Long Policy: Misogynistic vs Non-Misogynistic')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.tight_layout()

# Show the plot
plt.show()
```
