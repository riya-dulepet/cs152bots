```python
df1["true_label"] = "MISOGYNISTIC"
df2["true_label"] = "NON-MISOGYNISTIC"


import pandas as pd
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming df_misogynistic and df_non_misogynistic are your two DataFrames

# Combine both DataFrames
df_combined = pd.concat([df1[['true_label', 'label']], df2[['true_label', 'label']]])

# Create confusion matrix
cm = confusion_matrix(df_combined['true_label'], df_combined['label'], labels=['MISOGYNISTIC', 'NON-MISOGYNISTIC'])

# Plot the confusion matrix
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['MISOGYNISTIC', 'NON-MISOGYNISTIC'], yticklabels=['MISOGYNISTIC', 'NON-MISOGYNISTIC'])
plt.title('Gemini Long Policy: Misogynistic vs Non-Misogynistic')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

cm1 = np.array([[16, 14], [1, 29]])  # Gemini Long 
cm2 = np.array([[23, 7], [2, 28]])  # Gemini Short
cm3 = np.array([[17, 13], [0, 30]])   # OpenAI Long
cm4 = np.array([[26, 4], [6, 24]])  # OpenAI Short

fig, axes = plt.subplots(2, 2, figsize=(12, 12))
titles = ['Gemini Long Policy', 'Gemini Short Policy', 'OpenAI Long Policy', 'OpenAI Short Policy']

cms = [cm1, cm2, cm3, cm4]

for i, ax in enumerate(axes.flatten()):
    sns.heatmap(cms[i], annot=True, fmt='d', cmap='Blues', xticklabels=[ 'Misogynystic', 'Non-Misogynystic'],
                yticklabels=['Misogynystic', 'Non-Misogynystic'], ax=ax, cbar=False, annot_kws={'size': 24})
    ax.set_title(titles[i], fontsize=26)
    ax.set_xlabel('Predicted', fontsize=24)
    ax.set_ylabel('Actual', fontsize=24)
    ax.set_yticklabels(ax.get_yticklabels(), fontsize=20)
    ax.set_xticklabels(ax.get_yticklabels(), fontsize=20)

plt.tight_layout()
plt.show()
```

```python
# asked for help on how to make a double bar plot
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
severity_countsg = df_gemini['severity'].value_counts().sort_index()
severity_countso = df_openai['severity'].value_counts().sort_index()

# Align the indexes (ensures missing severity levels are included)
all_severity_levels = range(0, 5)
gemini_counts = severity_countsg.reindex(all_severity_levels, fill_value=0)
openai_counts = severity_countso.reindex(all_severity_levels, fill_value=0)

# Plot
x = np.arange(len(all_severity_levels))
width = 0.35

plt.figure(figsize=(10, 6))
plt.bar(x - width/2, gemini_counts, width, label='Gemini')
plt.bar(x + width/2, openai_counts, width, label='OpenAI')

plt.xlabel('Severity Level')
plt.ylabel('Number of Instances')
plt.title('Distribution of Severity Levels: Gemini vs. OpenAI')
plt.xticks(x, all_severity_levels)
plt.legend()
plt.tight_layout()
plt.show()
```
